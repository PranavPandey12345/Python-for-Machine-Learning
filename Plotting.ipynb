```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2,markersize=12)
plt.title("2d Diagram")
```

```python
#Creating Subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.plot(x,y)
```

```python
np.pi
```

```python
#Compute the x and y coordinaters for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
#Compute the x and y coordinaters for points on a sine and cosine currve
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set a subplot grid that has height 2 and width 1
#set the first such subplot as active
plt.subplot(2,1,1)

#Make the first plot
plt.plot(x,y_sin,'r--')
plt.title("Sine")

#Set the second subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title("Cosine")        
```

```python
#Bar plot

x=[2,8,10]
y=[11,16,9]

x1=[3,9,11]
y1=[6,15,7]
plt.bar(x,y)
plt.bar(x1,y1,color='g')
plt.title('Bar graph')
```

# Boxplot using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

# Pie Chart

```python
#Data to plot
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```

# Importing Seaborn Library

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Joinplot

Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

```python
#count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
#Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

# Boxplot

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```
